## 1.1 What Is Supermemory?

Supermemory is a **memory and context engine** for AI agents. It solves the "Goldfish Memory" problem -- AI systems forgetting everything between conversations. Think of it as a **living knowledge graph** -- not a static document store.

### The Problem It Solves
AI systems have "Goldfish Memory" -- every conversation starts from scratch. Users re-explain preferences, re-state project details, and re-introduce themselves.

### How It Works (Mental Model)

**Traditional Systems vs Supermemory:**
- Static files in folders -> Dynamic knowledge graph
- No connections -> Rich relationships (updates, extends, derives)
- Keyword search only -> Semantic understanding + keyword
- Information frozen -> Information evolves and connects
- Manual cleanup -> Automatic forgetting + contradiction resolution

### Core Capabilities

| Capability | What It Does |
|------------|-------------|
| **Memory Extraction** | Reads conversations/docs and extracts meaningful facts |
| **User Profiles** | Auto-builds static (long-term) and dynamic (recent) user context |
| **Hybrid Search** | Vector + keyword search with sub-300ms latency |
| **Graph Relationships** | Connects memories: updates, extends, derives |
| **Auto-Forgetting** | Expired/temporary facts decay automatically |
| **Connectors** | Google Drive, Notion, Gmail, OneDrive, GitHub sync |

## 1.4 Environment Setup

The SDK automatically reads `SUPERMEMORY_API_KEY` from your environment.
You can also pass it explicitly when creating the client.
Using an environment variable is best practice -- it keeps secrets out of source control.


````python
# ============================================================
# 1.5 Creating Your First Client
# ============================================================

import os
from supermemory import Supermemory

# Option 1: Let SDK auto-detect from environment
# (reads SUPERMEMORY_API_KEY env var automatically)
client = Supermemory()

# Option 2: Pass API key explicitly
# client = Supermemory(api_key="sm_your_key_here")

# Option 3: Point to self-hosted server
# client = Supermemory(
#     api_key="sm_local_key",
#     base_url="http://localhost:6767"
# )

print(f"Client ready!")
print(f"Base URL: {client.base_url}")
````

````python
# ============================================================
# 1.6 Verify Connectivity -- Add Your First Memory
# ============================================================

# Add a simple piece of content
response = client.add(
    content=(
        "Hello, Supermemory! My name is Alex, I'm a software engineer "
        "specializing in distributed systems. I love Python and Rust."
    ),
    container_tag="tutorial_user_alex"
)

print(f"Document ID: {response.id}")
print(f"Status: {response.status}")
````

````python
# ============================================================
# 1.7 Retrieve Context -- Get User Profile
# ============================================================

# The profile() method returns both static and dynamic facts
profile = client.profile(
    container_tag="tutorial_user_alex",
    q="What does Alex do?"  # Optional: search query for relevant memories
)

print("=== Static Profile (long-term facts) ===")
for fact in profile.profile.static:
    print(f"  - {fact}")

print()
print("=== Dynamic Profile (recent context) ===")
for fact in profile.profile.dynamic:
    print(f"  - {fact}")
````

## 1.8 How It Works Under the Hood

When you call `add()`:
1. Content is sent to the Supermemory engine
2. It queues the document for processing
3. The engine extracts semantic meaning, creates chunks, generates embeddings
4. Relationships are built between this memory and existing ones
5. The user profile is automatically updated

When you call `profile()`:
1. Static facts (long-term) are retrieved from the knowledge graph
2. Dynamic facts (recent context) are pulled from recent activity
3. If a query (`q`) is provided, relevant memories are searched and returned

**Key Insight:** Supermemory isn't just a vector database. It's a full context engine with a knowledge graph, user profiles, contradiction resolution, and intelligent forgetting.
