## 5.1 Not a Vector DB -- A Living Knowledge Graph

Traditional vector databases store chunks independently. Supermemory builds a **graph** where memories connect to other memories through typed relationships. It's facts built on top of other facts.

### Three Relationship Types

**1. UPDATES -- Information Changes**
When new info contradicts existing knowledge, the newer memory UPDATES the older one. The `isLatest` flag ensures searches return current information while preserving history.

**2. EXTENDS -- Information Enriches**
When new info adds detail without replacing, the newer memory EXTENDS the older one. Both remain valid -- searches get richer context.

**3. DERIVES -- Information Infers**
Supermemory infers new facts from patterns. These are insights you didn't explicitly state.


```python
# 5.2 Demonstrating 'Updates' -- contradiction resolution
from supermemory import Supermemory
client = Supermemory()

# Step 1: Add initial fact
client.add(
    content="Alex works at Google as a software engineer.",
    container_tag="user_alex"
)
print("Step 1: Alex works at Google")

# Step 2: Later, Alex changes jobs -- new info contradicts old
# Supermemory creates an UPDATE relationship automatically
client.add(
    content="Alex just started at Stripe as a Product Manager.",
    container_tag="user_alex"
)
print("Step 2: Alex now works at Stripe (UPDATES previous fact)")

# Step 3: Query -- returns CURRENT info, not old info
profile = client.profile(
    container_tag="user_alex",
    q="Where does Alex work?"
)
print("\n=== Profile after updates ===")
for fact in profile.profile.static:
    print(f"  - {fact}")
print("\nThe system tracked that Stripe is LATEST, Google is HISTORICAL.")
```

```python
# 5.3 Demonstrating 'Extends' -- adding richness
# Extends adds detail without replacing
client.add(
    content="Alex works at Stripe as a PM.",
    container_tag="user_alex"
)
client.add(
    content="Alex focuses on payments infrastructure and leads a team of 5 engineers.",
    container_tag="user_alex"
)
print("Added: Alex works at Stripe")
print("Added: Alex focuses on payments, leads team of 5")
print("\nBoth facts remain valid -- searches return enriched context.")
print("The second fact EXTENDS the first.")
```

```python
# 5.4 Demonstrating 'Derives' -- inferred insights
# Derives infers new facts from patterns
client.add(
    content="Alex is a PM at Stripe.",
    container_tag="user_alex"
)
client.add(
    content="Alex frequently discusses payment APIs and fraud detection.",
    container_tag="user_alex"
)
print("Fact 1: Alex is a PM at Stripe")
print("Fact 2: Alex discusses payment APIs and fraud detection")
print("\nSupermemory may DERIVE: Alex likely works on core payments.")
print("These inferences surface insights you never explicitly stated.")
```

```python
# 5.5 Automatic forgetting -- time-based decay
from datetime import datetime

# Temporary facts are automatically forgotten when expired
client.add(
    content=f"I have an exam tomorrow ({datetime.now().strftime('%Y-%m-%d')}).",
    container_tag="user_alex"
)
print("Added: I have an exam tomorrow")
print("After the exam date passes, this will be automatically forgotten.")

# Other auto-forgetting examples
print("\nExamples of auto-forgotten content:")
print("  - 'Meeting with Alex at 3pm today' -> forgotten after today")
print("  - 'I am at the airport' -> forgotten after some time")
print("  - Casual, non-meaningful content -> never becomes permanent")
```

```python
# 5.6 Memory types -- Facts, Preferences, Episodes
print("Supermemory distinguishes memory types automatically:")
print()
print("FACTS: Persist until updated")
print("  Example: 'Alex is a PM at Stripe'")
print()
print("PREFERENCES: Strengthen with repetition")
print("  Example: 'Alex prefers morning meetings'")
print("  Example: 'Alex likes functional programming'")
print()
print("EPISODES: Decay unless significant")
print("  Example: 'Met Alex for coffee Tuesday'")
print("  Example: 'Alex mentioned a new project idea'")
print()
print("All of this is AUTOMATIC. You don't tag types, clean up, or resolve contradictions.")
```

## 5.7 What You DON'T Need to Do

All of this is automatic. You NEVER need to:
- Define relationships manually
- Tag memory types
- Clean up old memories
- Resolve contradictions
- Manage knowledge versioning

Just add content and search naturally. The graph maintains itself.
