## 8.1 Why Self-Host?

- **Free, open source** -- no API key needed, no usage limits, no credit card
- **Privacy** -- your data never leaves your machine
- **Air-gapped** -- works fully offline with local models
- **Same API** -- drop-in compatible with the hosted platform
- **One binary** -- no Docker, no database to provision, no config files


```python
# 8.2 Installation -- one command
print("Install and start the server:")
print()
print("Option A: Install script")
print("  curl -fsSL https://supermemory.ai/install | bash")
print()
print("Option B: npx")
print("  npx supermemory local")
print()
print("On first boot:")
print("  - Embedded graph engine created automatically")
print("  - Built-in local embeddings (nothing sent externally)")
print("  - Auto-generated API key printed to console")
print("  - Full Memory API available at http://localhost:6767")
```

```python
# 8.3 Connecting SDK to local server (one line change)
from supermemory import Supermemory

# Point to local server
client = Supermemory(
    api_key="sm_local_key",  # printed on first boot
    base_url="http://localhost:6767"
)

print("Connected to local Supermemory server!")
print()
print("All the same methods work:")
print("  client.add(...)")
print("  client.profile(...)")
print("  client.search.documents(...)")
print("  client.documents.list(...)")
print()
print("AI coding plugins (Claude Code, Codex, OpenCode) work too:")
print('  export SUPERMEMORY_API_URL=http://localhost:6767')
```

```python
# 8.4 Bring your own model -- any LLM provider
print("Supermemory needs an LLM for memory extraction. You can bring any provider:")
print()
print("OpenAI:")
print('  OPENAI_API_KEY=sk-... OPENAI_MODEL=gpt-4o supermemory-server')
print()
print("Anthropic:")
print('  ANTHROPIC_API_KEY=sk-ant-... ANTHROPIC_MODEL=claude-sonnet-4-6 supermemory-server')
print()
print("Gemini:")
print('  GEMINI_API_KEY=... GEMINI_MODEL=gemini-2.5-pro supermemory-server')
print()
print("Groq:")
print('  GROQ_API_KEY=... GROQ_MODEL=llama-4-maverick supermemory-server')
print()
print("Any OpenAI-compatible endpoint works too:")
print('  OPENAI_BASE_URL=https://your-api.com/v1 OPENAI_API_KEY=... OPENAI_MODEL=... supermemory-server')
```

```python
# 8.5 Fully offline with Ollama (no internet needed after download)
print("For a fully offline setup:")
print()
print("1. Install Ollama: curl -fsSL https://ollama.ai/install.sh | sh")
print("2. Pull a model: ollama pull qwen3:32b")
print("3. Start Supermemory pointing to Ollama:")
print()
print('  OPENAI_BASE_URL=http://localhost:11434/v1 \\')
print('  OPENAI_API_KEY=ollama \\')
print('  OPENAI_MODEL=qwen3:32b \\')
print('  supermemory-server')
print()
print("Now: graph engine + embeddings + LLM are ALL local.")
print("Your data never leaves your machine.")
```

```python
# 8.6 Self-hosted vs Platform -- feature comparison
print("Feature Comparison:")
print()
print(f"{'Feature':<40} {'Self-hosted':<15} {'Platform':<15}")
print("-" * 70)
features = [
    ("Full Memory API", "Yes", "Yes"),
    ("Hybrid semantic search", "Yes", "Yes"),
    ("Local embeddings", "Your machine", "Managed"),
    ("File ingestion (PDFs, images)", "Yes", "Yes"),
    ("Connectors (GDrive, Notion, etc.)", "No", "Yes"),
    ("MCP Server", "No", "Yes"),
    ("Memory extraction quality", "Your model", "Proprietary (best)"),
    ("Cost", "Free", "Freemium"),
    ("Infrastructure", "Your machine", "Globally distributed"),
]
for name, self_h, plat in features:
    print(f"{name:<40} {self_h:<15} {plat:<15}")
print()
print("Best of both worlds: develop locally, deploy on platform with one base_url change.")
```

## 8.7 Key Takeaways

- Self-hosting is a single binary -- `curl ... | bash` and you're running
- Zero config: embedded database, auto-generated API key, built-in embeddings
- Bring any model: OpenAI, Anthropic, Gemini, Groq, or local with Ollama
- Fully offline possible: Ollama + Supermemory = local graph engine + local LLM
- SDK code is identical -- just change `base_url`
