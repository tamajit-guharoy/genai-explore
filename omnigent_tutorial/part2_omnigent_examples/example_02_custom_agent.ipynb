# Example 02: Writing Custom Agents

**Goal:** Write your own agent YAML files from scratch — different harnesses, tools, and os_env configurations.

**After this notebook:** You'll be able to define agents for coding, research, data analysis, and more.

## 1. The Anatomy of an Agent YAML

```python
# Key fields in every agent.yaml
required_fields = {
    'name': 'Stable identifier shown in sessions and logs',
    'prompt': 'System prompt (inline) — or use instructions: path/to/file',
    'executor': {
        'harness': 'claude-sdk | openai-agents | codex | cursor | pi | antigravity',
        'model': 'model identifier (provider-dependent)',
        'auth': 'Optional: databricks, api_key, provider'
    }
}

optional_fields = [
    'instructions',   # External prompt file (wins over prompt)
    'tools',          # MCP servers, Python functions, sub-agents
    'policies',       # Agent-level guardrails
    'os_env',         # OS tools access + sandbox config
    'terminals',      # Named interactive shells
    'async',          # Expose async work tools (default: true)
    'cancellable',    # Can session be cancelled (default: true)
    'timers',         # Expose timer tools (default: false)
    'spawn',          # Allow spawning child sessions
]

print("Required fields:")
for k, v in required_fields.items():
    print(f"  {k}: {v if isinstance(v, str) else '...'}")
print(f"\n{len(optional_fields)} optional fields")
```

## 2. Agent 1: A Simple Coding Agent (Claude)

```python
coding_agent = """
name: my_coder
description: A coding agent that writes tests first.

prompt: |
  You are a coding assistant who practices TDD (Test-Driven Development).
  For every task:
  1. Write the test first
  2. Run it (it should fail)
  3. Write the minimum code to pass
  4. Run tests again (should pass)
  5. Refactor if needed

executor:
  harness: claude-sdk
  model: claude-sonnet-4-6

os_env:
  type: caller_process
  cwd: .
  sandbox:
    type: none

async: true
cancellable: true
"""

with open('agent_coder.yaml', 'w') as f:
    f.write(coding_agent)
print("Created agent_coder.yaml")
print("Run: omni run agent_coder.yaml")
```

## 3. Agent 2: A Data Analyst (GPT via OpenAI Agents)

```python
analyst_agent = """
name: data_analyst
description: Analyzes CSV/JSON data and produces reports.

prompt: |
  You are a data analyst. When given data:
  1. First, describe the data: shape, columns, types, missing values
  2. Then, answer the specific question with statistical rigor
  3. Always include a summary table or chart description
  4. Flag any data quality issues you notice

executor:
  harness: openai-agents
  model: gpt-5

os_env:
  type: caller_process
  cwd: .
  sandbox:
    type: none
"""

with open('agent_analyst.yaml', 'w') as f:
    f.write(analyst_agent)
print("Created agent_analyst.yaml")
print("Run: omni run agent_analyst.yaml")
```

## 4. Agent 3: Research Agent with External Instructions

```python
# Create the instructions file
instructions_md = """# Research Agent Instructions

You are a thorough research assistant. For every query:

1. **Understand the question** — restate it to confirm
2. **Search comprehensively** — use web search, not just your training data
3. **Cite sources** — every factual claim gets a source link
4. **Acknowledge uncertainty** — be explicit about what you don't know
5. **Synthesize, don't list** — write a coherent answer, not bullet points
"""

with open('AGENTS.md', 'w') as f:
    f.write(instructions_md)

research_agent = """
name: researcher
description: Thorough research agent with web search.

instructions: AGENTS.md

executor:
  harness: claude-sdk
  model: claude-opus-4-8

os_env:
  type: caller_process
  cwd: .
  sandbox:
    type: none
"""

with open('agent_researcher.yaml', 'w') as f:
    f.write(research_agent)
print("Created AGENTS.md and agent_researcher.yaml")
print("instructions: takes precedence over any inline prompt:")
```

## 5. Harness Comparison Table

| Harness | Transport | Best For | CLI Needed? |
|---|---|---|---|
| `claude-sdk` | Anthropic Python SDK | Coding, reasoning | No |
| `openai-agents` | OpenAI Python SDK | GPT-native workflows | No |
| `claude-native` | PTY (tmux) | Full Claude Code experience | Yes (`claude`) |
| `codex-native` | PTY (tmux) | Full Codex CLI experience | Yes (`codex`) |
| `codex` | OpenAI Python SDK | GPT with tool use | No |
| `pi` | PTY or SDK | Multi-model, exploration | Yes (`pi`) for native |
| `cursor` | Proprietary | Cursor editor integration | Yes (`cursor-agent`) |
| `antigravity` | Google Antigravity SDK | Gemini models | No |
