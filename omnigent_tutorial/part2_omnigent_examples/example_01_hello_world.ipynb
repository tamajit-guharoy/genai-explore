# Example 01: Hello, Omnigent

**Goal:** Install Omnigent, set up credentials, and launch your first agent.

**Prerequisites:** Python 3.12+, `uv` installed. Node.js 22+ and npm recommended.

## 1. Check Your Environment

```python
import sys
print(f"Python version: {sys.version}")
assert sys.version_info >= (3, 12), "Need Python 3.12+"
print("✓ Python 3.12+ confirmed")
```

```python
import subprocess, shutil

# Check uv
uv_path = shutil.which('uv')
print(f"uv path: {uv_path}")

# Check node
node_path = shutil.which('node')
print(f"node path: {node_path}")

# Check tmux (Linux/macOS only)
tmux_path = shutil.which('tmux')
print(f"tmux path: {tmux_path or 'NOT FOUND (OK on Windows)'}")
```

## 2. Install Omnigent

If not already installed, run this cell:

```python
# Install Omnigent via uv (run this in a terminal if not yet installed)
# !uv tool install omnigent

# Verify installation
import subprocess
try:
    result = subprocess.run(['omni', '--version'], capture_output=True, text=True)
    print(f"Omnigent version: {result.stdout.strip()}")
except FileNotFoundError:
    print("Omnigent not installed yet. Run: uv tool install omnigent")
```

## 3. Set Up Credentials

Omnigent auto-detects API keys from your environment. Check what's available:

```python
import os

providers = {
    'ANTHROPIC_API_KEY': 'Anthropic (Claude)',
    'OPENAI_API_KEY': 'OpenAI (GPT)',
    'GEMINI_API_KEY': 'Google (Gemini)',
    'CURSOR_API_KEY': 'Cursor',
    'DEEPSEEK_API_KEY': 'DeepSeek',
}

for var, name in providers.items():
    status = '✓ FOUND' if os.environ.get(var) else '✗ not set'
    print(f"  {name:25s} {status}")

print()
print("Run 'omni setup' for interactive credential setup if needed.")
```

## 4. Test a Basic Agent

Create a minimal agent and run it:

```python
# Write a minimal agent YAML
agent_yaml = """
name: hello_agent
prompt: |
  You are a concise assistant. Answer questions directly.

executor:
  harness: claude-sdk
  model: claude-sonnet-4-6
"""

agent_path = 'hello_agent.yaml'
with open(agent_path, 'w') as f:
    f.write(agent_yaml)

print(f"Created {agent_path}")
print("Run with: omni run hello_agent.yaml")
```

## 5. Omnigent Concepts Demonstrated

| Concept | What you did |

| **Agent as YAML** | Defined an agent in a single file |
| **Harness abstraction** | Chose `claude-sdk` — could swap to `openai-agents` |
