## 8.1 Why Evaluate Agents?

LLMs are non-deterministic — the same input can produce different outputs.
For production agents, you need systematic testing:

| Test Type | What it verifies |
|-----------|-----------------|
| **Factuality** | Agent returns correct information |
| **Safety** | Agent refuses harmful requests |
| **Tool use** | Agent uses the right tool for the right task |
| **Honesty** | Agent says "I don't know" instead of fabricating |
| **Format** | Agent follows output format instructions |

## 8.2 Deployment Options

| Option | Command | Best For |
|--------|---------|----------|
| **Cloud Run** | `adk deploy cloud_run my_agent/` | Serverless, auto-scaling |
| **Vertex AI Agent Engine** | `adk deploy agent_engine my_agent/` | Enterprise, managed |
| **GKE** | `adk deploy gke my_agent/` | Kubernetes, custom networking |
| **Docker** | `adk build docker my_agent/` | Portable, on-prem, hybrid |
| **API Server** | `adk api_server my_agent/` | Local dev, integration testing |

## Production Checklist

Before deploying an agent to production:
- [ ] Specific, well-structured instructions (ROLE/TOOLS/CONSTRAINTS/OUTPUT)
- [ ] Safety callbacks (block dangerous tool calls)
- [ ] Eval cases covering key scenarios
- [ ] Monitoring (Cloud Trace + Cloud Logging)
- [ ] API keys stored in Secret Manager (not .env files!)
- [ ] Rate limiting and authentication on the endpoint
